````python
from string import punctuation

def contar_palavra(palavra):
    
    livro = open('quincas.txt').read()
    livro = livro.lower()
    livro = livro.replace('\n', ' ')
    livro_no_punctuation = ""
    
    for i in livro:
        if i not in punctuation:
            livro_no_punctuation = livro_no_punctuation + i
    
    lista_livro = livro_no_punctuation.split(" ")

    contador = 0
    for i in lista_livro:
        if palavra == i:
            contador += 1
    
    return contador
````

````python
print(contar_palavra('cachorro'))

print(contar_palavra('machado'))

print(contar_palavra('cachorrão'))
````

## While loops

While loops verificação determinada condição e rodam um bloco de código *enquanto* a condição é verdadeira.


### Exemplo 1

````python
condicao = True
contador = 0

while condicao:
    contador = contador + 1
    print('Quantidade de vezes que o loop rodou:', contador)
    
    if contador == 10:
        condicao = False    
````

### Exemplo 2

Simulador de execução de pena:

````python
pena_cumprida = 0
tempo_condenado = 5
contador_anos = 0

while pena_cumprida < tempo_condenado:
    contador_anos = contador_anos + 1

    print("Tempo total (sem descontos) para o condenado ser solto: ", tempo_condenado - pena_cumprida)
    
    print("Tempo líquido: ", contador_anos)
    
    comportamento_exemplar = input("O Prisoneiro teve comportamento exemplar? (reponder com Sim ou Não)")
    
    if comportamento_exemplar == "Sim":
        pena_cumprida = pena_cumprida + 2
    else:
        pena_cumprida = pena_cumprida + 1
````

# Orientação a objetos

Vimos anteriormente que praticamente tudo em python são objetos --- objetos do tipo string, do tipo list, do tipo dict e etc. Mas de onde vêm os objetos?

## Classes

Todos os objetos tem uma classe associada --- que corresponde ao seu tipo. Objetos do tipo string são da classe *string*, listas são da classe lista e por ai vai.

Pensem nas classes como "moldes" para os objetos --- a partir delas criamos os objetos. Na maior parte das vezes, elas estão "escondidas", no sentido que não a vemos e simplesmente utilizamos os objetos.

### Sintaxe das classes

**Não se preocupem em entender exatamente como escrever as classes, mas antes a lógica por trás da "orientação a objetos", ou seja, o porque da existência dos métodos e características de cada classe.**




````python
class Dog():

    def __init__(self, nome):
        self.nome = nome
        self.mamifero = True
        self.patas = 4

    def latir(self):
        print('Au Au!')
        
    def imprimir_nome(self):
        print(self.nome)
````

Podemos criar um objeto a partir a classe da seguinte maneira:


````python
lila = Dog('lila')
````

Quando rodamos ```Dog('lila')```, a **função padrão** \_\_init\_\_ é executada automaticamente. Essa função tem o papel de inicializar determinados elementos da classes. Todas as variáveis que ela recebe são colocadas diretamente na criação do objeto.

As demais funções são o que chamamos anteriormente de *métodos*. Por exemplo, podemos chamar a função ```latir()```, que é um método da classe ```Dog()``` --- todos os objeto criados a partir dessa classe terão esse método.

Temos, também, os **atributos** associados às classes. Assim como os métodos são funções embutidas, **atributos** são **objetos embutidos**. 

````python
lila.latir()
````

````python
print(lila.mamifero)

print(lila.nome)
````
